```python
from megaparse import MegaParse
from megaparse.configs.auto import MegaParseConfig
from langchain_openai import ChatOpenAI

MegaParseConfig()

megaparse = MegaParse()
response = megaparse.load(DOC_LIST[0])
print(response)
```
